```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("time.txt",sep=" ",header=None,names=["Start","End","No_of_cores","Time_Taken"])
```

```python

df_r1=df[df["End"]==500000]
```

```python
df["No_of_cores"].head()
```

```python
x=df_r1["No_of_cores"]
y=df_r1["Time_Taken"]
```

```python
plt.plot(x,y)
plt.xlabel("No_of_cores")
plt.ylabel("Time_Taken")
plt.text(0.2,0.8,"RANGE IS 1 - 500000",transform=plt.gca().transAxes)
```
