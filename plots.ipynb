```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("timeparallel.txt",sep=",",header=None,names=["N_value","NO_OF_CHILDREN","TIME(MS)"])
```

```python
df.head()
```

```python
df_r1=df[df["N_value"]==500]
x=df_r1["NO_OF_CHILDREN"]
y=df_r1["TIME(MS)"]
```

```python
plt.plot(x,y)
plt.xlabel("NO_OF_CHILDREN")
plt.ylabel("Time_Taken")
plt.text(0.2,0.8,"Size of n is 500",transform=plt.gca().transAxes)
```
